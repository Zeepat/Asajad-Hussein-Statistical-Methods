# Centrala Gränsvärdessatsen

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import norm # Importera normalfördelning från scipy.stats

iter = 5000 # Antal iterationer
means = np.zeros(iter) # Skapa en tom array för medelvärdena
n_list = [2, 3, 4, 20, 50] # Skapa en lista med värden på n

fig = plt.subplots() # Skapa tom figur

for n in n_list: # Loopa över alla värden på n

    for i in range(iter): # Den här loopoen medelvärdesbildar så många gånger som vi bestämt i 'iter'.
        rv = norm.rvs(size=n) # Skapa en normalfördelning med storlek n
        mean = rv.mean() # Beräkna medelvärdet
        means[i] = mean # Spara medelvärdet i arrayen 'means'

    fig = sns.kdeplot(means)

fig.legend(n_list).set_title('n (# samples)') # Skapa en legend med värdena på n
```

```python
from scipy.stats import expon # Importera exponentialfördelning från scipy.stats

iter = 5000 # Antal iterationer
means = np.zeros(iter) # Skapa en tom array för medelvärdena
n_list = [2, 3, 4, 20, 50] # Skapa en lista med värden på n

for n in n_list: # Loopa över alla värden på n

    for i in range(iter): # Den här loopoen medelvärdesbildar så många gånger som vi bestämt i 'iter'.
        rv = expon.rvs(size=n) # Skapa en exponentialfördelning med storlek n
        mean = rv.mean() # Beräkna medelvärdet
        means[i] = mean # Spara medelvärdet i arrayen 'means'

    fig = sns.kdeplot(means)

fig.legend(n_list).set_title('n (# samples)') # Skapa en legend med värdena på n
```

```python
from scipy.stats import binom # Importera exponentialfördelning från scipy.stats

iter = 5000 # Antal iterationer
means = np.zeros(iter) # Skapa en tom array för medelvärdena
n_list = [2, 3, 4, 20, 50] # Skapa en lista med värden på n

fig = plt.subplots() # Skapa tom figur

for n in n_list: # Loopa över alla värden på n

    for i in range(iter): # Den här loopoen medelvärdesbildar så många gånger som vi bestämt i 'iter'.
        rv = binom.rvs(n=5, p=0.5, size=n) # Skapa en binomialfördelning med storlek n
        mean = rv.mean() # Beräkna medelvärdet
        means[i] = mean # Spara medelvärdet i arrayen 'means'

    fig = sns.kdeplot(means)

fig.legend(n_list).set_title('n (# samples)') # Skapa en legend med värdena på n
```

```python
from scipy.stats import uniform # Importera exponentialfördelning från scipy.stats

iter = 5000 # Antal iterationer
means = np.zeros(iter) # Skapa en tom array för medelvärdena
n_list = [2, 3, 4, 20, 50] # Skapa en lista med värden på n

fig = plt.subplots() # Skapa tom figur

for n in n_list: # Loopa över alla värden på n

    for i in range(iter): # Den här loopoen medelvärdesbildar så många gånger som vi bestämt i 'iter'.
        rv = uniform.rvs(size=n) # Skapa en binomialfördelning med storlek n
        mean = rv.mean() # Beräkna medelvärdet
        means[i] = mean # Spara medelvärdet i arrayen 'means'

    fig = sns.kdeplot(means)

fig.legend(n_list).set_title('n (# samples)') # Skapa en legend med värdena på n
```

```python
from scipy.stats import t # Importera t-fördelningen från SciPy

iter = 5000 # Antalet iterationer
means = np.zeros(iter) # Skapa en tom NumPy-array med plats för alla våra medelvärden
n_list = [2, 3, 5, 20, 50] # Skapa en lista med värden på n

fig, ax = plt.subplots() # Skapa tom figur

for n in n_list: # Loopa över en lista med olika värden på n
    
    #5
    for i in range(iter): # Den här loopen medelvärdesbildar så många gånger som vi bestämt i 'iter'.
        rv = t.rvs(df=2, size=n) # Tag n slumpmässiga värden från t-fördelningen (1 frihetsgrad) med t.rvs()
        mean = rv.mean() # Bilda medelvärdet av dem
        means[i] = mean # Spara medelvärdet i listan means
        
    #6
    fig = sns.kdeplot(means) # Plotta medelvärdena med en KDE-plot i Seaborn
#7
ax = ax.set(xlim=(-5, 5)) # Sätt gränsvärden för x-axel
fig.legend(n_list).set_title('n (# samples)') # Lägg till legend med värde på 'n'
```
